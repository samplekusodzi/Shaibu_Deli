### Object Oriented Programing


```python
### We want to create an interactive dashboard which will ask clients for their booking details
### This is to help the transport sector manage customer demands
```

```python
class TravelPlan:
    buses =  ["VIP","STC","OA","Metro Mass"] ## global variable 

    def welcome_user(self): # this is a method
        print("You are welcome to ABC travel agency")
        
    
    def user_destination(self):
        destination = input("Enter your destination")
        print(f"You want to go to {destination}")

    def select_bus(self):
       print("\nAvailable Buses:") 
       for index, bus in enumerate(self.buses, start=1):
           print(f"{index}:{bus}")
       choice = int(input("Select your preferred bus"))
       actual_bus = self.buses[choice-1]
       print(f"You selected: {actual_bus}")
       return actual_bus
    
    def date_travel(self):
        date = input("Enter you travel date")      
```

```python
### creating objects or instance
plan = TravelPlan()
plan.welcome_user()
plan.user_destination()
plan.select_bus()
plan.date_travel()
```
